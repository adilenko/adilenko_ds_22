2) Критикам нравятся больше игры на PC или на PS4?
3) Критикам больше нравятся стрелялки или стратегии?

```python

import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats as st
```

```python

df = pd.read_csv('vgsales.csv')
df.head(100)
```

```python
df_s = df.loc[df.Genre == 'Sports']["Critic_Score"].dropna()
print(df_s)
df.loc[df.Genre == 'Sports']["Critic_Score"].mean()
```

```python
#H0 Критики хорошо относятся к спортивным играм, средняя оценка >= 70
#H1 Критики не очень хорошо относятся к спортивным играм, средняя оценка < 70

a=0.05
result = st.ttest_1samp(df_s, 70)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < a):
    print('Отвергаем нулевую гипотезу, среднее меньше 70')
else:
    print('Не отвергаем нулевую гипотезу')
# одновыборочный, односторонний критерий    
# Разница значимая но  статистика больше нуля.     
```

 Критикам нравятся больше игры на PC или на PS4

```python
#H0 Средняя оценка критиков для PC > средней оценки для PS4
#H1 Средняя оценка критиков для PC < средней оценки для PS4

df_pc = df.loc[df.Platform == 'PC']["Critic_Score"].dropna()
df_ps4 = df.loc[df.Platform == 'PS4']["Critic_Score"].dropna()
print(df_pc.mean())
print(df_ps4.mean())
a = 0.05


result = st.ttest_ind(df_pc, df_ps4, equal_var=False)
print(result)
if (result.pvalue / 2 < a):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
    
# двухвыборочный, односторонний критерий 
# разница не значима не можем отвергнуть нулевую гипотизу
```

Критикам больше нравятся стрелялки или стратегии

```python
#H0 Средняя оценка критиков для стрелялок больше средней оценки для стратегии 
#H1 Средняя оценка критиков для стрелялок меньше средней  оценки для стратегии 

df_shooter = df.loc[df.Genre == 'Shooter']["Critic_Score"].dropna()
df_strateg = df.loc[df.Genre == 'Strategy']["Critic_Score"].dropna()
print(df_shooter.mean())
print(df_strateg.mean())
a = 0.05


result = st.ttest_ind(df_shooter, df_strateg, equal_var=False)
print(result)
if (result.pvalue / 2 < a):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
    
 # двухвыборочный, односторонний критерий   
# pvalue меньше уровня значимости, statistic < 0 значит отвергаем нулевую гепотизу
# Получаем что средняя оценка критиков для стрелялок меньше средней  оценки для стратегии
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
wordnet_lemmatizer = WordNetLemmatizer()
nltk.download('stopwords')
nltk.download('wordnet')
spam_orig = pd.read_csv('spam.csv')
spam = spam_orig.copy()
spam.head(10)
spam.info()
```

```python
spam["Message"] = spam["Message"].apply(lambda x: x.lower())
spam["Message"] = spam["Message"].apply(lambda x: re.sub('[\W_]+',' ', x))
stopwords_set = set(stopwords.words('english'))
spam["Message"] = spam["Message"].apply(lambda x: [word for word in x.split() if word not in stopwords_set])
spam["Message"] = spam["Message"].apply(lambda x: " ".join([wordnet_lemmatizer.lemmatize(word) for word in x])) 

tfidf_matrix = tfidf.fit_transform(spam.Message)
print(tfidf_matrix)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)


```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, spam.Category, test_size=0.30, random_state=42)
lr = LogisticRegression(random_state=42)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
accuracy_score(y_test, y_pred)
#модель предсказывает значения с хоршей точностью = 96%
```

```python
confusion_matrix(y_test, y_pred, labels=["ham", "spam"])
#в тестовой выборке 3  хам сообщения были приняты за спам и 66 спам сообщений было принято за хам
```
