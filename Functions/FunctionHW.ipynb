```python
commands = ["p", "s", "l", "as", "ds", "ad", "d", "q"]
documents = [
    {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
    {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
    {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
]

directories = {
    "1": ["2207 876234", "11-2"],
    "2": ["10006"],
    "3": []
}


def checker(legal_commands):
    def decorator(func):
        def wrapper(*args, **kwargs):
            if args[0] in legal_commands:
                try:
                    print(func(*args, **kwargs))
                except ValueError as err:
                    print(err)
            else:
                print("Command is not valide:")
            return None

        return wrapper

    return decorator


def get_owner():
    num = input("Enter document number: ")
    msg = ""
    owner_dict = [el for el in documents if el["number"] == num]
    if owner_dict:
        msg = f"The document owner is {owner_dict[0]['name']}"
    else:
        raise ValueError("There is no such document")
    return msg


def get_shelf():
    msg = ""
    num = input("Enter document number: ")
    shelf = dict(filter(lambda el: num in el[1], directories.items()))
    if shelf:
        msg = f"Document shell is  {list(shelf)[0]}"
    else:
        raise ValueError("There is no such document")
    return msg


def get_all_info():
    msg = "Documents list: \n"
    nl = '\n'
    for doc in documents:
        shelf, _ = list(filter(lambda el: doc["number"] in el[1], directories.items()))[0]
        msg += f"#{doc['number']} Type: {doc['type']} Name: {doc['name']} Shelf: {shelf}{nl}"
    return msg


def add_shelf():
    msg = ""
    num = input("Enter shelf number: ")
    if num not in directories:
        directories[num] = []
        msg = f"Shelf is added. Availabel shelfs {', '.join(list(directories.keys()))}"
    else:
        msg = f"{num} shelf already exists"
    return msg


def delete_shelf():
    msg = ""
    num = input("Enter shelf number: ")
    if num not in directories:
        raise ValueError(f"{num} shelf doesn't exist. Availabel shelfs {', '.join(list(directories.keys()))}")
    elif num in directories and directories[num]:
        msg = f"The shelf is not empty. Remove documents first. Availabel shelfs {', '.join(list(directories.keys()))}"
    elif num in directories and not directories[num]:
        directories.pop(num)
        msg = f"The shelf is removed. Availabel shelfs {', '.join(list(directories.keys()))}"
    return msg


def add_document():
    doc_num = input("Enter document number: ")
    doc_type = input("Enter document type: ")
    doc_owner = input("Enter document owner: ")
    shelf_num = input("Enter shelf number: ")
    if shelf_num not in  directories:
        raise ValueError(f"{shelf_num} shelf doesn't exist. Availabel shelfs {', '.join(list(directories.keys()))}")
    documents.append({"type": doc_type, "number": doc_num, "name": doc_owner})
    directories[shelf_num].append(doc_num)
    return get_all_info()


def remove_document():
    doc_num = input("Enter document number: ")
    rem_ind = None
    for i, el in enumerate(documents):
        if doc_num == el["number"]:
            rem_ind = i
    if rem_ind is None:
        raise ValueError("There is no such document")
    else:
        del(documents[i])
        for d in directories.values():
            if doc_num in d:
                d.remove(doc_num)
    return get_all_info()


@checker(commands)
def command_handler(command):
    msg = ""
    if command == "p":
        msg = get_owner()
    elif command == "s":
        msg = get_shelf()
    elif command == "l":
        msg = get_all_info()
    elif command == "as":
        msg = add_shelf()
    elif command == "ds":
        msg = delete_shelf()
    elif command == "ad":
        msg = add_document()
    elif command == "d":
        msg = remove_document()
    elif command == "q":
        msg = "Bye"
    return msg


def main():
    com = ""
    while com != "q":
        com = input("Enter a command: ")
        command_handler(com)


main()
```
