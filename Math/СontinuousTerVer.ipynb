```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import pandas as pd
import math
```

Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
args =[{"mu":1,"sigma":0.5}, {"mu":5,"sigma":0.5},{"mu":5,"sigma":1},{"mu":5,"sigma":2},{"mu":5,"sigma":3}]
plt.figure(figsize=(20, 10)) 
for arg in args:
    norm_rv = sts.norm(loc=arg["mu"], scale=arg["sigma"])
    x = np.linspace(-2, 10, 1000)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf)
plt.show()

```

```python
lams = [0, 1,2,3,4,5]

for lam in lams:
    expon_rv = sts.expon(lam)
    x = np.linspace(0, 5, 1000)
    pdf = expon_rv.pdf(x)
    plt.plot(x, pdf)

plt.show()
```

```python

N=[1,2,10,50,100]
for n in N:
    t_rv = sts.t(n)
    x = np.linspace(-5, 5, 1000)
    pdf = t_rv.pdf(x)
    plt.plot(x, pdf)

plt.show()
```
