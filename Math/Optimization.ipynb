```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import dual_annealing
from scipy.optimize import differential_evolution

def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

def differential_calculation(**kwargs):
    res = differential_evolution(ackley, bounds, seed=42, **kwargs)
    res = {"function_value":res.fun, "nit":res.nit, "x":res.x, "success":res.success, **kwargs}
    print(res)
    return res

bounds = [(-10, 10), (-10, 10)]
strategies = ["best1bin", "best1exp","rand1exp","randtobest1exp","currenttobest1exp","best2exp","rand2exp","randtobest1bin",
         "currenttobest1bin","best2bin","rand2bin","rand1bin"]
result = differential_evolution(ackley, bounds, seed=42)
print("Default result", result)

```

```python
x = strategies
results=[differential_calculation(strategy=strategy) for strategy in strategies]
fig = plt.figure(figsize=(20, 5))
ax = fig.add_axes([0,0,1,1])
fig.suptitle('Strategy', fontsize=20)
ax.bar(x,[res["nit"] for res in results])
plt.show()
# в данном случае наименьшее количество итерации при bes1bin стратегии
```

```python
x = list(range(1,20))
results=[differential_calculation(popsize=n) for n in range(1,20)]
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes([0,0,1,1])
fig.suptitle('popsize', fontsize=20)
ax.bar(x,[res["nit"] for res in results])
plt.show()
#при малой популяции количество итерации уменьшается, но не находит глобальный минимум, при popsiz =5 начинаем находить 
#глобальный минимум
```

```python
import random
x = [0, 0.02, 0.04, 0.06, 0.08, 0.1, 0.15, 0.2, 0.3, 0.4, 0.5, 0.8, 1]
results=[differential_calculation(tol=n) for n in x]
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes([0,0,1,1])
fig.suptitle('tol', fontsize=20)
ax.bar(x,[res["nit"] for res in results])
plt.show()
# начиная со значения толерантности 0.3 < tol <= 0.4  перестаём находить глобальный минимум
```

```python

x = [z/10 for z in range(0,20,1)]
results=[differential_calculation(mutation=n) for n in x]
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes([0,0,1,1])
fig.suptitle('mutation', fontsize=20)
ax.bar(x,[res["nit"] for res in results])
plt.show()

x = [z/100 for z in range(10,30,1)]
results=[differential_calculation(mutation=n) for n in x]
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes([0,0,1,1])
fig.suptitle('mutation', fontsize=20)
ax.bar(x,[res["nit"] for res in results])
plt.show()
#похоже с точностью до сотых лучший результат nit: 23, получаем при mutation = 0.12
```

```python
x = [p/10 for p in range(0,10,1)]
results=[differential_calculation(recombination=n) for n in x]
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes([0,0,1,1])
fig.suptitle('recombination', fontsize=20)
ax.bar(x,[res["nit"] for res in results])
plt.show()
#похоже увеличения значения вероятности в общем случае ведёт к уменьшения кол-ва итерации. 
#Лучшее значение по кол-ву итераций получил при 'recombination': 0.8
```

```python
x = ["immediate", "deferred"]
results=[differential_calculation(updating=val) for val in x]
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes([0,0,1,1])
fig.suptitle('updating', fontsize=20)
ax.bar(x,[res["nit"] for res in results])
plt.show()
#при updating = immediate кол-во итераций меньше
```
