Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.

```python
from scipy.misc import derivative
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x+3)
x=1
f_p1 = derivative(f, x, dx=1e-6)
print("Первая производная для x=1", f_p1)

x= symbols('x')
f = x**5 + 4*sin(2*x) + cos(3*x+3)
f_pr1 = diff(f,x)
print("Первая производная", f_pr1)
f_pr2= diff(f_pr1,x)
print("Вторая производная", f_pr2)
f_pr2_val = f_pr2.evalf(subs={x: 1})
print("Вторая производная для x=1", f_pr2_val)

```


Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python

x= symbols('x')
f = sin(2*x+1)**5
f_pr1 = diff(f,x)
print("Первая производная", f_pr1)
f_pr2= diff(f_pr1,x)
print("Вторая производная", f_pr2)
x_ar = np.linspace(-5, 5)

fy = [f.evalf(subs={x:xx}) for xx in x_ar]
f_pr1y = [f_pr1.evalf(subs={x:xx}) for xx in x_ar]
f_pr2y = [f_pr2.evalf(subs={x:xx}) for xx in x_ar]

plt.plot(x_ar, fy, label='function')
plt.plot(x_ar, f_pr1y, label='derivative')
plt.plot(x_ar, f_pr2y, label='second derivative')
plt.legend()
plt.show()
```
