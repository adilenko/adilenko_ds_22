```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
lan = 0.1
rv = sts.expon(lan)
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()


```

```python
#параметры запуска: размеры выборок и количество повторов генерации
n_l = [2 ,5, 10, 30, 50, 100, 200, 500]
samples_count = 1000
plt.ylabel('samples means')
plt.xlabel('$x$')
for n in n_l:
    a1 = []
    for number in range(samples_count):
        sample1 = rv.rvs(size=n)
        a1.append(sample1.mean())
    label_1 = 'sample of ' + str(n) + ' histogram'    
    plt.hist(a1, density=True, label=label_1, bins=30)
    plt.legend()
    plt.show()
```
